```python
import gensim
import numpy as np
```

```python
def title_vec(title, w2v_model):
	title = title.split(' ')
	vector = np.zeros(300)
	for word in title:
		try:
			vector += w2v_model[word]
		except:
			pass
	return vector
```

```python
w2v_model = gensim.models.KeyedVectors.load_word2vec_format('../GoogleNews-vectors-negative300.bin', binary=True)
a=title_vec("I love elephants",w2v_model)
```
